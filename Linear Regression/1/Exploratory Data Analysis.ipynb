```python
"""
The Boston housing dataset is our sample dataset that provides median values of homes in different areas around Boston. 
Along with the median home values in $1000 (MEDV), 
crime (CRIM), 
nitric oxides concentration (NOX),
the average number of rooms (RM), 
percentage of lower status of the population (LSTAT), 
and other features are provided

Our objective is to predict median home price (MEDV)
"""

from sklearn.datasets import load_boston
from sklearn.datasets import fetch_california_housing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

```

```python
california_housing_dataset = fetch_california_housing()

california_housing = pd.DataFrame(california_housing_dataset.data, columns=california_housing_dataset.feature_names)

#We then add the target into the DataFrame
california_housing['MEDV'] = california_housing_dataset.target

```

```python
"""
If the DataFrame contains more than just numeric values, 
by default, describe() outputs the descriptive statistics for the numeric columns. 
To show the summary statistics of all column, specify include = 'all' in the method
"""

california_housing.describe(include = 'all').round(2)
```

```python
"""
Summary statistics provides a general idea of each feature and the target, 
but visualization reveals the information more clearly. 

It’s a good practice to visualize and inspect the distribution column by column.
"""

california_housing.hist(column='AveRooms',bins=20,figsize=(15,5))
plt.show()

california_housing['AveRooms'].plot(kind='hist',ylim=(0,25000),bins=100)
plt.show()

california_housing.hist(column='AveRooms', bins=20)
plt.show()

"""
The distribution of AveRooms appears normal and symmetric. 
The symmetry aligns with what we observed from the output of describe()

Informative data visualization not only reveals insights, but they are invaluable to communicate findings to stakeholders.
"""
```
